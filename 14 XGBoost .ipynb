# XGBoost

- XGBoost, which stands for Extreme Gradient Boosting, is a machine learning algorithm that excels at handling structured data, particularly for tasks like regression and classification. 
- It belongs to the category of ensemble learning, which combines the strengths of multiple models to create a more robust and accurate final model.

```python
X = df.drop('Channel', axis=1) #indepent

y = df['Channel'] #dependent
```

```python
# convert labels into binary values

y[y == 2] = 0

y[y == 1] = 1
```

```python
# import XGBoost
import xgboost as xgb


# define data_dmatrix
data_dmatrix = xgb.DMatrix(data=X,label=y)
```

```python
# split X and y into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# import XGBClassifier
from xgboost import XGBClassifier


# declare parameters
params = {
            'objective':'binary:logistic',
            'max_depth': 4,
            'alpha': 10,
            'learning_rate': 1.0,
            'n_estimators':100
        }
            
            
            
# instantiate the classifier 
xgb_clf = XGBClassifier(**params)



# fit the classifier to the training data
xgb_clf.fit(X_train, y_train)
```

```python
# alternatively view the parameters of the xgb trained model
print(xgb_clf)
```

```python
# make predictions on test data
y_pred = xgb_clf.predict(X_test)
```

```python
# check accuracy score
from sklearn.metrics import accuracy_score

print('XGBoost model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```
